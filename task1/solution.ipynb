```python
import pandas as pd

df = pd.read_csv('task-1-stocks.csv', index_col=False)
df = df.rename_axis(index=None, columns=None)
df
```

```python
cov_df = df.pct_change().cov()
```

```python
c = df.to_numpy()
n = len(c[0]) # число акций
t = len(c) # длина временого ряда
cov = cov_df.to_numpy()
```

```python
import numpy as np
m = int(np.log2((1e6 / np.median(c[0]))))
```

```python
import numpy as np

r = np.array([])
sigma = np.array([])
for i in range(len(c[0])):
    p = c[:, i]
    r_s = [(p[j + 1] - p[j]) / p[j] for j in range(t - 1)]
    r_ = np.mean(r_s)
    r = np.append(r, r_)
    sigma = np.append(sigma, np.sqrt(len(r_s) * sum([((r_s[j] - r_) ** 2) / (len(r_s) - 1) for j in range(len(r_s))])))

print(r)
print(sigma)
```

Примем обозначения <br>
$r_i$ - средняя доходность i-го актива <br>
$\sigma_i$ - уровень риска i-го актива <br>
$x_i$ - число акций i-го актива <br>
$c_i$ - цена i-го актива в певый день <br>
$N$ - число активов

Запишем $x_i$ в бинарной форме (с шагом 50 для уменьшения числа переменных)<br>
$x_i = 50 * \sum_{k=0}^{M} 2^k y_k$ <br>
$y_k = \begin{cases}1, \text{если мы взяли акцию в портфель} \\ 0, \text{в ином случае} \end{cases}$

**Формулировка задачи <br>**
Максимизировать доходность $max \sum_{i=1}^{N} r_i x_i$ <br>
Поддержитвать уровень риска s. t. $\sum_{i=1}^{N} \sigma_i x_i = 0.2$ <br>
Стоимость портфеля $\sum_{i=1}^{N} c_i x_i = 1.000.000$ <br>

Тогда задачу можно свести к QUBO <br>
$f_Q = f_A + f_B$ <br>
$f_A = A (10^6 - \sum_{i=1}^{N} c_i x_i)^2 + A (0.2 - \sum_{i=1}^{N} \sigma_i x_i)^2$ определяет ограничения задачи (сумма и рискованность портфеля соответственно) <br>
$f_B = -B \sum_{i=1}^{N} r_i x_i$ это целевая функция задачи <br>

А матрицу QUBO тогда можно записать в виде <br>
$Q = \begin{cases}
A \cdot (c_i^2 \cdot \alpha ^ 2 - 2 \cdot c_i \cdot \alpha \cdot 10 ^ 6) + A \cdot (\sigma_i^2 \cdot \alpha ^ 2 - 2 \cdot \alpha \cdot \sigma_i \cdot 0.2) - B \cdot r_i \cdot \alpha,\text{ }i = j \\ 
A \cdot c_i \cdot \alpha \cdot c_j \cdot \alpha + A \cdot \sigma_i \cdot \alpha \cdot \sigma_j \cdot \alpha, \text{ }i \neq j \end{cases}$ <br>
$\alpha = 50 * 2^k$

```python
Q = np.array([[0] * n * m] * n * m)
print(Q.shape)
A = 1
B = 0.5
for i in range(n):
    for j in range(n):
            for k in range(m):
                alpha = 2 ** (k - 1)
                if i * m + k <= j * m + k:
                    # if i * m + k == j * m + k:
                    #     Q[i * m + k][j * m + k] = A * (cov[i, j] + (c[0, i] ** 2 * alpha ** 2 - 2 * c[0, i] * alpha * 1e6)) - B * r[i] * alpha
                    # else:
                    #     Q[i * 8 + k][j * m + k] = A * alpha ** 2 * (c[0, i] * c[0, j] + cov[i, j])
                    if i * m + k == j * m + k:
                        Q[i * m + k][j * m + k] = (A * ((c[0, i] ** 2 + sigma[i] ** 2) * alpha ** 2 - 2 * (c[0, i] * 1e6 + sigma[i] * 0.2) * alpha) 
                                               - B * r[i] * alpha)
                    else:
                        Q[i * m + k][j * m + k] = A * alpha ** 2 * (c[0, i] * c[0, j] + sigma[i] * sigma[j])
```

```python
import dimod

Q = dimod.BQM(Q, "BINARY")
```

```python
# import warnings
import neal
from neal import SimulatedAnnealingSampler

# warnings.filterwarnings("ignore", category=DeprecationWarning)
sampler = SimulatedAnnealingSampler()
sampleset = sampler.sample(Q, num_reads=100)
best_solution = sampleset.first.sample
```

```python
numbers = []
for i in range(n):
    val = 0
    for k in range(m):
        val += int(best_solution[i * m + k]) * 2 ** k
    numbers.append(val)

numbers
```

```python
print("Сумма закупки: ", sum([c[0, i] * numbers[i] for i in range(n)]))
print("Доходность: ", sum([r[i] * numbers[i] / sum(numbers) for i in range(n)]))
print("Уровень риска: ", sum([sigma[i] * numbers[i] for i in range(n)]) / sum(numbers))
```
